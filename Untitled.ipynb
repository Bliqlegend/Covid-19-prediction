```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
recovered_cases=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
confirmed_cases=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
cases_country=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/web-data/data/cases_country.csv')
```

```python
confirmed_cases=confirmed_cases.replace(np.nan,' ',regex=True)
cases_country=cases_country.replace(np.nan,' ',regex=True)
deaths=deaths.replace(np.nan,' ',regex=True)
recovered_cases=recovered_cases.replace(np.nan,' ',regex=True)
```

```python
country='India'
confirmed_india=confirmed_cases[confirmed_cases['Country/Region']==country]
confirmed_india=confirmed_india.drop(["Province/State",'Lat','Long','Country/Region'],axis=1)

deaths_india=deaths[deaths['Country/Region']==country]
deaths_india=deaths_india.drop(["Province/State",'Lat','Long','Country/Region'],axis=1)

recovered_india=recovered_cases[recovered_cases['Country/Region']==country]
recovered_india=recovered_india.drop(["Province/State",'Lat','Long','Country/Region'],axis=1)
confirmed_india.columns
```

```python
confirmed_india_ts=confirmed_india.reset_index(drop=True).sum()
recovered_india_ts=recovered_india.reset_index(drop=True).sum()
deaths_india_ts=deaths_india.reset_index(drop=True).sum()
active_india_ts=pd.Series(np.array([x1-x2-x3 for (x1,x2,x3) in zip(confirmed_india_ts,recovered_india_ts,deaths_india_ts)]),index=confirmed_india_ts.index)
```

```python
india_array=[confirmed_india_ts,recovered_india_ts,deaths_india_ts,active_india_ts]
labels=["Confirmed",'Recovered','Deaths','Active']
color_arr=px.colors.qualitative.Dark24
```

```python
#draw_plot(india_array)
#plt.plot(confirmed_india_ts,color='red',marker='o',linewidth=1,markersize=5)
#plt.plot(recovered_india_ts)
#plt.plot(deaths_india_ts)
#plt.plot(active_india_ts)
fig_1=go.Figure(data=go.Scatter(x=confirmed_india_ts.index,y=confirmed_india_ts.values,mode='lines+markers'))
fig_1.update_layout(title="Total Confirmed cases (globally)",yaxis_title='Confirmed Cases',xaxis_tickangle=315)
fig_1.show()
```

```python
fig=go.Figure()
pos=0
for i in india_array:
    fig.add_trace(go.Scatter(x=i.index,y=i.values,name=labels[pos]))
    pos+=1
fig.update_layout(title="Total Cases (India)",xaxis_title='Cases from January to September',yaxis_title='Total Cases',xaxis_tickangle=315)
fig.show()
```
